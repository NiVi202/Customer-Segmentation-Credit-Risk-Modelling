```python
df.head()
```

```python
X=df
```

```python
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
```

```python
X
```

```python
X.plot(subplots=True, grid=False, title="After Standardizatio",
        layout=(9,2), legend=True,x_compat=True,
        figsize=(20,30))
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(X_std)
arr=np.cumsum(pca.explained_variance_ratio_)
plt.plot(arr)
plt.xlim(0,18,1)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
X1=[0,2.8]
Y1=[0.6,0.6]
X2=[0,4]
Y2=[0.7,0.7]
X3=[0,3]
Y3=[0.63,0.63]
X4=[0,6]
Y4=[0.82,0.82]
X5=[6,6]
Y5=[0,0.82]
plt.plot(X1,Y1,X2,Y2,X3,Y3,X4,Y4,X5,Y5)
plt.show()
```

```python
from sklearn.decomposition import PCA 
sklearn_pca = PCA(n_components=6)
Y_sklearn = sklearn_pca.fit_transform(X_std)
```

```python
Y_sklearn.shape
```

```python
Y_sklearn
```

```python
arr1=pd.DataFrame(Y_sklearn)
arr1
```

```python
from sklearn.cluster import KMeans
```

```python
Sum_of_squared_distances = []
K = range(1,20)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(Y_sklearn)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
XA=[0,5]
YA=[60000,60000]
XB=[5,5]
YB=[0,60000]
plt.plot(XA,YA,XB,YB)
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(Y_sklearn)
```

```python
print(kmeans.cluster_centers_)
```

```python
arr=kmeans.labels_
group=pd.DataFrame(arr,columns=['clust'])
group
```

```python
group.describe()
```

```python
label = pd.concat((group,arr1),axis=1)
label.rename(columns = {0:'PC_1'}, inplace = True)
label.rename(columns = {1:'PC_2'}, inplace = True)
label.rename(columns = {2:'PC_3'}, inplace = True)
label.rename(columns = {3:'PC_4'}, inplace = True)
label.rename(columns = {4:'PC_5'}, inplace = True)
label.rename(columns = {5:'PC_6'}, inplace = True)
```
