```python
from sklearn.preprocessing import StandardScaler
df_std = StandardScaler().fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(df_std)
k=range(1,17)
plt.plot(np.cumsum(pca.explained_variance_ratio_),color="red")
plt.xlim(0,17,1)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
#Through Scikit learn we are plotting graph for no. of components over our data after data cleaning.
```

```python
from sklearn.decomposition import PCA 
sklearn_pca = PCA(n_components=7) #90% of variance is covered around 7 components from the above plot.
Y_sklearn = sklearn_pca.fit_transform(df_std)
```

```python
print(Y_sklearn)
```

```python
Y_sklearn.shape
```

```python
df=pd.DataFrame(Y_sklearn,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7'])
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
wcss = [] 
m = range(1,11)
for i in m: 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df) 
    wcss.append(kmeans.inertia_)
```

```python
#elbow method for finding optimal no. of K
plt.plot(m,wcss,'bx-')
plt.plot(range(1, 11), wcss,color="red")
plt.xlabel('No. of Principle components')
plt.ylabel('Inertia') 
plt.show()
```

```python
X = df[["PC1", "PC2","PC3","PC4","PC5","PC6","PC7"]]
plt.scatter(X["PC1"], X["PC2"], c="r")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
K=3 #through elbow method

Centroids = (X.sample(n=K))
plt.scatter(X["PC1"], X["PC2"], c="cyan")
plt.scatter(Centroids["PC1"], Centroids["PC2"], c="r") #getting three centroids as K=3
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
