```python
from sklearn.cluster import KMeans
wcss = []
m = range(1,11)
for i in m:
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
    
#Finnding the optimum no of clusters(K) using the elbow method
```

```python
plt.plot(m,wcss,'bx-')
plt.plot(range(1,11),wcss)
plt.xlabel('PC')
plt.ylabel('Inertia')
plt.show()

#The K is the reading on which the slope in the changes the most in the graph below 
```

```python
K=3

Centroids = (X.sample(n=K))
plt.scatter(X["PC1"], X["PC2"], c="green")
plt.scatter(Centroids["PC1"], Centroids["PC2"], c="red")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()

#Now, knowing the number of clusers to make, we choose any three points to begin with clustering.
```

```python
Centroids
```

```python
diff = 1
j=0

while(diff!=0):
    XD=X
    i=1
    for index1, row_c in Centroids.iterrows():
        ED=[]
        for index2, row_d in XD.iterrows():
            d1 = (row_c["PC1"]-row_d["PC1"])**2
            d2 = (row_c["PC2"]-row_d["PC2"])**2
            d3 = (row_c["PC3"]-row_d["PC3"])**2
            d4 = (row_c["PC4"]-row_d["PC4"])**2
            d5 = (row_c["PC5"]-row_d["PC5"])**2
            d6 = (row_c["PC6"]-row_d["PC6"])**2
            d7 = (row_c["PC7"]-row_d["PC7"])**2
            d = sqrt(d1+d2+d3+d4+d5+d6+d7)
            ED.append(d)
        X[i] = ED
        i = i+1
    
    C = []
    for index, row in X.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(K):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos = i+1
        C.append(pos)
    X["Cluster"]=C
    Centroids_new = X.groupby(["Cluster"]).mean()[["PC1", "PC2","PC3","PC4","PC5","PC6","PC7"]]
    if j == 0:
        diff = 1
        j = j+1
    else:
        diff = (Centroids_new['PC1'] - Centroids['PC1']).sum() + (Centroids_new['PC2'] - Centroids['PC2']).sum() + (Centroids_new['PC3'] - Centroids['PC3']).sum() + (Centroids_new['PC4'] - Centroids['PC4']).sum() +(Centroids_new['PC5'] - Centroids['PC5']).sum()+ (Centroids_new['PC6'] - Centroids['PC6']).sum()+(Centroids_new['PC7'] - Centroids['PC7']).sum()
        print(diff.sum())
    Centroids = X.groupby(["Cluster"]).mean()[["PC1","PC2","PC3","PC4","PC5","PC6","PC7"]]
    
    #we canculate the distance of all the points from those three points and assign each one to one of those three initial points.
    #We then calculate the centroid of the 3 groups so formed. We again do the similar process for these three centroids and keep doing it untill th centroids do not change. 
```
